## Min Pooling function

```python
#takes too long
def minfilter(img,kernel):
    '''
    img is an nd array (grayscale)
    kernel is binary filter
    '''
    kernel_size = int(kernel.size**0.5)
    padding = int(kernel_size/2)
    max_val = np.max(img)
    shape = img.shape
    x=shape[1]
    y=shape[0]
    padded_img = np.concatenate((np.ones((padding,x+(2*padding)))*max_val, np.concatenate((np.ones((y,padding))*max_val,img, np.ones((y,padding))*max_val),axis=1) , np.ones((padding,x+2*(padding)))*max_val), axis=0)
    output = img
    for i in range(padding,x+padding):
        for j in range(padding,y+padding):
            crop = padded_img[j-padding:j+padding+1,i-padding:i+padding+1]
            output[j-padding,i-padding] = np.min(crop * kernel)
    
    return output
    
```
